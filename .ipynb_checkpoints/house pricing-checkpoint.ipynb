```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as stats

import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)
```

```python
house = pd.read_csv("train_house.csv")
```

```python
house.head ()
```

```python
house.tail ()
```

```python
print(house.isnull().sum())
```

```python
house = house.dropna(axis=1, thresh=house.shape[0] * 0.5)
```

```python
print(house.shape)
print(house.head())
```

```python
house_numerical = house.select_dtypes(include=['int64', 'float64'])
```

```python
print("Numerical Data:")
print(house_numerical.head())
```

```python
print(house_numerical.isnull().sum())
```

```python
house_numerical["LotFrontage"].fillna(house_numerical["LotFrontage"].median(), inplace=True)
house_numerical["MasVnrArea"].fillna(house_numerical["MasVnrArea"].median(), inplace=True)
house_numerical["GarageYrBlt"].fillna(house_numerical["GarageYrBlt"].median(), inplace=True)
```

```python
print(house_numerical.isnull().sum().sum())
```

```python
print(house_numerical.describe())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(house_numerical["SalePrice"], kde=True, bins=30)
plt.title("Distribusi Harga Rumah")
plt.show()
```
